```python
# convert the model to C

import m2cgen as m2c
import joblib

def generateModelCode(model_file, source_file):
    model = joblib.load(model_file)
    code = m2c.export_to_c(model)
    with open(source_file, 'w') as f:
        f.write(code)

generateModelCode("Debts-Hyperdrive-bestmodel.pkl","model.c")
```

```python
# generate the fields description values for the embedded model - categoric values encoding and max/min values

import pandas as pd
df_ = pd.read_csv('Debts.csv',sep=';',low_memory=False)
classname="RESULTADO"

# first correct null values

cat_features=[]
for col, col_type in df_.dtypes.iteritems():
    if col != classname:
        if col_type == 'O':
            cat_features.append(col)
            df_[col].fillna(df_[col].value_counts().index[0], inplace=True)  # categoricos: preenche nulos com valor mais frequente
        else:
            df_[col].fillna(df_[col].median(), inplace=True)  # numéricos: preenche nulos com média

def dumpField(col,col_type):
    print("=============================")
    if col != classname:
        print("Attribute: " + col + " - Type: " + str(col_type))
    else:
        print("Class: " + col + " - Type: " + str(col_type))
    if col_type == 'O':
        df_[col] = df_[col].astype('category')
        print("Values:")
        values = dict(enumerate(df_[col].cat.categories))
        for key in values:
            print(str(key) + " - " + str(values[key]))
    else:
        print("Min: " + str(df_[col].min()) + " - Max: " + str(df_[col].max()))
        
for col, col_type in df_.dtypes.iteritems():
    dumpField(col,col_type)
```
